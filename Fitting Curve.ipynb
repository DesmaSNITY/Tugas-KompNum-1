untuk mengimpor library seperti matplotlib untuk membuat visualisasi grafik, dan numpy untuk pengolahan data numerik. Baris %matplotlib inline memastikan bahwa grafik ditampilkan langsung di notebook.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

Data acak dibuat menggunakan numpy dengan generator angka acak.

  Data x adalah 50 angka acak antara 0 dan 10.
  
  Data y dihitung berdasarkan formula y = 2x - 1 dengan menambahkan sedikit
  noise acak untuk membuat data lebih realistis.
  
  Scatter plot dibuat untuk menampilkan hubungan antara x dan y.

```python
rng = np.random.RandomState(0)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y);
```

untuk membuat dan melatih model regresi linear, memprediksi nilai berdasarkan model tersebut, dan menampilkan hasilnya dalam bentuk grafik. Kode ini bekerja dengan data yang sudah ada (x dan y) untuk mencocokkan garis linear terbaik.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
model.fit(x[:, np.newaxis], y)
xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])
plt.scatter(x, y)
plt.plot(xfit, yfit);
```

model.coef_: Mengembalikan slope (kemiringan) dari garis regresi, yang dalam hal ini mendekati nilai 2.

model.intercept_: Mengembalikan intercept (titik potong dengan sumbu-y), yang mendekati nilai -1.

```python
print("Model slope : ",model.coef_[0])
print("Model Intercept : ",model.intercept_)
```
